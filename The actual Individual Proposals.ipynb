1. Does Talking to People Increase Happiness?

**Research Question**: *Does general social interaction make people happier?*

- Population Parameter: The average happiness score for each level of interaction frequency to see if more frequent interactions correlate with higher happiness.

**Variables**:
- Independent: Frequency of social interactions:
  - `CONNECTION_activities_talked_day_p3m`: Daily interactions.
  - `CONNECTION_activities_talked_family_p3m`: Family interactions.
  - `CONNECTION_activities_talked_job_p3m`: Work-related interactions.
  - `CONNECTION_activities_talked_hobbies_p3m`: Hobby-related conversations.

- Dependent: Happiness score (`WELLNESS_subjective_happiness_scale_happy`), rated from 1 (not happy) to 7 (very happy).

**Visualization**: A bar plot with error bars to display mean happiness scores for each interaction frequency, showing variability with standard error.

**Analysis Plan**:

I want to:
1. Calculate the mean happiness score for each frequency level.
2. Use standard error to visualize variability.

**Hypothesis**: Higher social interaction frequency such as daily or weekly, will correlate with higher happiness scores.

Relevance: If frequent social interaction is linked to higher happiness, it suggests that promoting regular social engagement could improve well-being.


2. Does More Sleep Reduce Tiredness?

**Research Question**: *Does getting “just the right amount” of sleep reduce tiredness?*

- Population Parameter: The average frequency of feeling tired across sleep amount categories to assess if “just right” sleep relates to less tiredness.

**Variables**:
- Independent: Perceived sleep amount (`LIFESTYLE_time_use_balance_sleeping`), such as:
  - “Too much”
  - “Too little”
  - “Just the right amount”

- Dependent: Tiredness frequency (`WELLNESS_malach_pines_burnout_measure_tired`), with options like "Always," "Very Often," "Rarely," etc.

**Visualization**: A bar plot with error bars showing the mean tiredness score for each sleep category.

**Analysis Plan**:
I want to:
1. Calculate the average tiredness score for each sleep category.
2. Use standard error bars to display variability.

**Hypothesis**: Individuals with “just the right amount” of sleep will report lower tiredness than those with “too much” or “too little” sleep.

Relevance: Supports balanced sleep for reducing tiredness, potentially useful for public health messaging.


3. Does Commute Time Affect Life Satisfaction?

**Research Question**: *Does longer commute time decrease life satisfaction?*

- Population Parameter: The average life satisfaction score for each commute category to examine if shorter commutes correlate with higher life satisfaction.

**Variables**:
- Independent: Commute time (`GEO_commute_time`), categorized as:
  - "No Commute"
  - "Less than 15 minutes"
  - "15 to 29 minutes"
  - "30 to 44 minutes"
  - "45 to 59 minutes"
  - "60 minutes and over"

- Dependent: Life satisfaction (`WELLNESS_malach_pines_burnout_measure_tired`), recorded in levels like "Always," "Very Often," "Sometimes," etc.

**Visualization**: A bar plot with error bars to compare average life satisfaction across commute times.

**Analysis Plan**:

I want to:
1. Calculate the mean life satisfaction score for each commute category.
2. Use standard error to illustrate variability.

**Hypothesis**: Shorter commutes are associated with higher life satisfaction.

Relevance: If shorter commutes lead to higher life satisfaction, this could support policies promoting remote work or reducing commute times for better well-being.
